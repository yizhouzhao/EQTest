```python
import os
import json
```

```python
root_folder = "E:/Temp/TestImages"
```

```python
img_lst = []
```

```python
img_lst
```

```python
for folder_name in os.listdir(root_folder):
    img_folder = root_folder + "/" + folder_name
    img_info = [[],[],[]]
    for file_name in os.listdir(img_folder):
        if file_name.startswith("persp"):
            img_info[0].append(img_folder + "/" + file_name)
        elif file_name.startswith("camera1"):
            img_info[1].append(img_folder + "/" + file_name)
        else:
            img_info[2].append(img_folder + "/" + file_name)
    img_lst.append([folder_name, img_info])
```

```python
with open("E:/Temp/jsonURL.txt", 'w') as outfile:
    json.dump(img_lst, outfile)
```
